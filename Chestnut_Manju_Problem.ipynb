```python
import math
import matplotlib.pyplot as plt

def calculate_time(initial_volume, target_volume, doubling_time):
    if initial_volume >= target_volume:
        return 0
    ratio = target_volume / initial_volume
    n_doublings = math.ceil(math.log(ratio, 2))
    return n_doublings * doubling_time

def plot_growth(initial_volume, target_volume, doubling_time):
    total_time = calculate_time(initial_volume, target_volume, doubling_time)
    times = list(range(0, total_time + doubling_time, doubling_time))
    volumes = [initial_volume * (2 ** (t / doubling_time)) for t in times]

    plt.figure(figsize=(10, 6))
    plt.plot(times, volumes, 'b-', label='Chestnut Manju Volume')
    plt.axhline(y=target_volume, color='r', linestyle='--', label='Solar System Volume')
    plt.yscale('log')
    plt.xlabel('Time (minutes)')
    plt.ylabel('Volume (m³)')
    plt.title('Exponential Growth of Chestnut Manju')
    plt.legend()
    plt.grid(True)
    plt.show()

# Parameters
manju_diameter = 0.05  # 5 cm in meters
manju_radius = manju_diameter / 2
manju_volume = (4/3) * math.pi * manju_radius**3

solar_system_radius = 4.5e12
solar_system_volume = (4/3) * math.pi * solar_system_radius**3

doubling_time = 5  # minutes

# Results
time_minutes = calculate_time(manju_volume, solar_system_volume, doubling_time)
print(f"Time to cover the Solar System: {time_minutes} minutes")

plot_growth(manju_volume, solar_system_volume, doubling_time)
```
